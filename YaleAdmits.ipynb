 Question 1:Over the past thirty years, what trends have changed in the number of applicants and admissions to Yale?
 

```python
##Group the data by year and then calculate the total number of applicants and admissions for each year
admission_summary = Yale_Admissions_df.groupby('Year Entered')[['Applications', 'Admits']].sum()
```

```python
import matplotlib.pyplot as plt
```

```python
##Generate a bar chart
admission_summary.plot(kind='bar', figsize=(16, 12))
plt.xlabel('Year Entered')
plt.ylabel('Count')
plt.title('Applications vs Admits Over the Years')
plt.legend(['Applications', 'Admits'])
plt.show()
```

Conclusion

I visualized the relationship between the year, the number of applicants, and the number of admissions in the above graph, because I want to explore whether the number of admissions will also increase or decrease as the number of applicants increases or decreases. I used Yale University's enrollment statistics from 1986 to 2022 to display the annual number of applicants and admissions in the form of a bar chart. From this chart, it can be seen that the number of applicants at Yale University does not affect the number of admissions. It seems to have shown a steady growth trend in the past thirty years, especially in 2020 and 2021, when there was a sharp increase.

Question 2:Do the number of applicants and admissions vary more in certain regions or countries?

```python
##Set the 'Year Entered' column as an index
Yale_Admissions_df.set_index('Year Entered', inplace=True)
```

```python
# Select region or country related columns
location_columns = ['NewHaven', 'NewEngland', 'NewYorkState','EastofMississippi','WestofMississippi','U.S.Poss.Foreign']
location_data = Yale_Admissions_df[location_columns]
```

```python
# Draw a line chart
location_data.plot(kind='line', figsize=(10, 6))
plt.xlabel('Year Entered')
plt.ylabel('Number of Admits')
plt.title('Admissions Comparison in Different Locations')
plt.legend(location_columns)
plt.show()
```

Conclusion

I visualized the changes in the number of admissions over time in different regions (such as New Haven, New England, New York State, etc.) using the above chart. Because I want to analyze Yale University's regional enrollment policy and the changing trends in different regions.From the graph, it can be seen that U S. Pos and foreign regions have always accounted for only a small proportion of admissions, and admissions have remained stable, indicating that Yale University's admission standards for international students may be relatively strict, or the number of applicants in these regions may be relatively small. The number of admissions in the East of Mississippi region continues to be higher than in other regions. This may be due to the higher overall quality of applicants in the region, meeting Yale University's admission criteria. In the past few years, the number of admissions in New York State has shown a downward trend. This may reflect a decrease in the number of applicants in the region or an overall quality of applicants that does not meet Yale University's admission criteria. In 2020, except for the number of international students admitted, there was a significant decrease in the number of admissions in other regions. This may be due to the impact of the global pandemic, resulting in a decrease in the number of applications or adjustments in enrollment policies. Based on these trends, it can be inferred that Yale University's admission strategy in different regions may be influenced by various factors such as regional characteristics, the quality and quantity of applicants.
